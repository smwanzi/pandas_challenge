# PyCitySchools Analysis

Based on the data analysis:

•	PyCitySchools district performed above average with an overall passing rate of 80.39% with students doing better in reading (81.88%) compared to math’s (78.99). This suggest more attention should be put towards improving performance in math.

•	The top 5 schools were Charter Schools, with a passing rate of 95.20% - 95.59% and the bottom 5 schools were District Schools, with a passing rate of 73.29% - 73.80%, this shows a significant difference in performance between Charter and District schools. One explanation for this could be that the district schools had on average many students.

•	Small (94.82%) and Medium (95.20%) sized schools performed better than the Large (76.36%) sized schools, suggesting a correlation between the size of the school and performance. 

•	schools that spent <585 dollars per student had a passing rate of 95.04%, whereas those that spent >645 dollars had and overall passing rate of 73.65.  This suggests the spending more per student does not significantly lead to better performance.

In conclusion, I would recommend to the district to have more charter schools, reduce the number of students in district schools and maintain a budget of 615 dollars per student.



```python
# Dependencies
import pandas as pd
import numpy as np
```

```python
# Load the the academy of Py school data files
schools = 'Resources/schools_complete.csv'
students = 'Resources/students_complete.csv'
```

```python
# open and read the schools and students data
# format decimals
pd.options.display.float_format = '{:,.2f}'.format
schools_df = pd.read_csv(schools)
students_df = pd.read_csv(students)
#students_df.head()
```

```python
# rename columns 
schools = schools_df.rename(columns={"school_name": "school"})
students = students_df.rename(columns={"student_name": "student", "school_name": "school"})
```

```python
#convert budget and size data type to integer for easier data manipulation
schools ["budget"] = pd.to_numeric(schools ["budget"])
schools ["size"] = pd.to_numeric(schools ["size"])
```

```python
# merge schools and students datasets into one dataframe and assign a variable
pyacademy_df = pd.merge(schools,students,how="left", on="school")
#convert score / size / budget data type to integer
pyacademy_df["reading_score"] = pd.to_numeric(pyacademy_df["reading_score"])
pyacademy_df["math_score"] = pd.to_numeric(pyacademy_df["math_score"])
pyacademy_df["size"] = pd.to_numeric(pyacademy_df["size"])
pyacademy_df["budget"] = pd.to_numeric(pyacademy_df["budget"])
#pyacademy_df.head()
```

## Create a high level snapshot (in table form) of the district's key metrics

```python
# Calculate the total number of schools, students, budget from the schools' dataframe
total_schools = schools_df["School ID"].count()
total_students = students_df["Student ID"].count()
total_budget = schools_df["budget"].sum()
```

```python
# Calculate Average Math Score,Reading Score, and percentage passing scores
average_math = students_df['math_score'].mean()
average_reading = students_df['reading_score'].mean()
pass_math = (students_df['math_score']>=70).sum()/total_students
pass_reading = (students_df['reading_score']>=70).sum()/total_students
pass_overall = (pass_math+pass_reading)/2
```

# District Summary


```python
# Display District Summary with a cleaner formatting, floats to 2 decimal & budget with $ sign.
district_summary_df = pd.DataFrame({
    "Total Schools":[total_schools],
    "Total Budget":[total_budget],
    "Total Students":[total_students],
    "Avg Math Score":[average_math],
    "Avg Reading Score":[average_reading],
    "Percentage Passing Math":[pass_math*100],
    "Percentage Passing Reading":[pass_reading*100],
    "Percentage Passing Overall":[pass_overall*100]})
district_summary_df['Total Budget'] = district_summary_df['Total Budget'].map('${:,.2f}'.format)
district_summary_df
```

```python
# for the summary key metrics table, group pyacademy_df data by school name 

grouped_school = pyacademy_df.set_index('school').groupby(['school'])

school_type = schools.set_index('school')['type']

total_student = grouped_school ['Student ID'].count()

school_budget = schools.set_index('school')['budget']

student_budget = schools.set_index('school')['budget']/schools.set_index('school')['size']

avg_math = grouped_school ['math_score'].mean()
avg_read = grouped_school ['reading_score'].mean()

# % passing
pass_math = pyacademy_df [pyacademy_df ['math_score'] >= 70].groupby('school')['Student ID'].count()/total_student* 100

pass_read = pyacademy_df [pyacademy_df ['reading_score'] >= 70].groupby('school')['Student ID'].count()/total_student * 100

overall_passing = (pass_math + pass_read) / 2

school_summary_df = pd.DataFrame({
    "School Type": school_type,
    "Total Students": total_student,
    "Total School Budget": school_budget,
    "Per Student Budget": student_budget,    
    "Average Math Score": avg_math,
    "Average Reading Score": avg_read,
    '% Passing Math': pass_math,
    '% Passing Reading': pass_read,
    "Overall Passing Rate": overall_passing
})
# format table
school_summary_df["Total School Budget"] = school_summary_df["Total Students"].map("${:,.2f}".format)
school_summary_df["Per Student Budget"] =school_summary_df["Per Student Budget"].map("${:,.2f}".format)
school_summary_df
```
